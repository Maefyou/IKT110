## non linear weights

```python
import matplotlib.pyplot as plt
import numpy as np

w1 = [1,-1]
b1 = [-1,1]
w2 = [1,1]
b2 = [0]

def relu(x):
    return max(0,x)

def forward(x):
    h1 = relu(w1[0]*x + b1[0]), relu(w1[1]*x + b1[1])
    h2 = w2[0]*h1[0] + w2[1]*h1[1] + b2[0]
    return h2

x = [i/10 for i in range(0,20)]
y = [forward(i) for i in x]
plt.plot(x,y)
```
